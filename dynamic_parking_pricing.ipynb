**1. Importing and Preprocessing Data**

```python
df = pd.read_csv("dataset.csv")
df
```

```python
# Combine the 'LastUpdatedDate' and 'LastUpdatedTime' columns into a single datetime column
df["Timestamp"] = pd.to_datetime(df["LastUpdatedDate"] + " " + df["LastUpdatedTime"], format="%d-%m-%Y %H:%M:%S")
df.head()
```

```python
#Ensure occupancy <= capacity
df = df[df['Occupancy'] <= df['Capacity']]
df
```

```python
#Use manual ordinal encoding to convert VehicleType values to numerical values
vehicle_map = {
    "truck": 1.5,
    "car": 1.0,
    "bike": 0.7,
    "cycle": 0.5
}

df["VehicleType"] = df["VehicleType"].str.strip().str.lower()
df["VehicleType_weighted"] = df["VehicleType"].map(vehicle_map)
df.head()
```

```python
#Use manual ordinal encoding to convert TrafficConditionNearby to numerical value
traffic_condition_map = {
    "low": 0,
    "average": 1,
    "high": 2
}

df["TrafficConditionNearby"] = df["TrafficConditionNearby"].str.strip().str.lower()
df["traffic_condition_weighted"] = df["TrafficConditionNearby"].map(traffic_condition_map)
df.head()
```

```python
#Convert SystemCodeNumber to numerical value so parking lot number can be seen easily

#Put all 14 unique parking lot names into array
parking_arr = df.SystemCodeNumber.unique()

#Function to initialize a map that maps parking lot name to number
def initialize_lot_map(lot_names):
  parking_lot_map = {}
  for i in range(0, 14):
    lot_id = lot_names[i]
    parking_lot_map[lot_id] = i

  return parking_lot_map

print(initialize_lot_map(parking_arr))
```

```python
#Now we use this map to create the encoded parking_lot_number column
df["parking_lot_number"] = df["SystemCodeNumber"].map(initialize_lot_map(parking_arr))
df.head()
```

```python
# Define the schema for the streaming data using Pathway
# This schema specifies the expected structure of each data row in the stream

class ParkingSchema(pw.Schema):
  parking_lot_number: int
  Timestamp: str
  Occupancy: int
  VehicleType_weighted: float
  traffic_condition_weighted: int
  QueueLength: int
  IsSpecialDay: int
  Capacity: int
  Latitude: float
  Longitude: float
```

```python
# Activate Panel (if not already done)
#pn.extension()
```

```python
# Load the data as a simulated stream using Pathway's replay_csv function
# This replays the CSV data at a controlled input rate to mimic real-time streaming
# input_rate=1000 means approximately 1000 rows per second will be ingested into the stream.

data = pw.demo.replay_csv(
      "parking_stream.csv",
      schema=ParkingSchema,
      input_rate=100
    )
```

```python
# Define the datetime format to parse the 'Timestamp' column
fmt = "%Y-%m-%d %H:%M:%S"

# Add new columns to the data stream:
# - 't' contains the parsed full datetime
# - 'day' extracts the date part and resets the time to midnight (useful for day-level aggregations)
data_with_time = data.with_columns(
    t = data.Timestamp.dt.strptime(fmt),
    day = data.Timestamp.dt.strptime(fmt).dt.strftime("%Y-%m-%dT00:00:00")
)
```

**2. Model 1 - Simple Pricing Function**

```python
import datetime

# Step 1: Daily tumbling window over time
enhanced_daily_window = (
    data_with_time.windowby(
        pw.this.t,
        instance=(pw.this.day, pw.this.parking_lot_number),  # group by calendar day
        #instance = pw.this.day,
        window=pw.temporal.tumbling(datetime.timedelta(days=1)),
        behavior=pw.temporal.exactly_once_behavior()
    )
    .reduce(
        t = pw.this._pw_window_end,
        lot = pw.reducers.max(pw.this.parking_lot_number),
        occ_sum = pw.reducers.sum(pw.this.Occupancy),
        cap = pw.reducers.max(pw.this.Capacity),
        count = pw.reducers.count(),
    )
)
```

```python
# Step 2: Compute averages
enhanced_daily_window = enhanced_daily_window.with_columns(
    occ_avg = pw.this.occ_sum / pw.this.count
)

enhanced_daily_window = enhanced_daily_window.with_columns(
    occupancy_ratio = pw.this.occ_avg / pw.this.cap
)
```

```python
#Step 3: Compute unclamped price from ratio (alpha=5)
enhanced_daily_window = enhanced_daily_window.with_columns(
    raw_price = 10 + 5 * pw.this.occupancy_ratio  # α = 5
)
```

```python
#Step 4: Clamp final price between $5-$20
enhanced_daily_window = enhanced_daily_window.with_columns(
    price = pw.if_else(
        pw.this.raw_price < 5, 5,
        pw.if_else(pw.this.raw_price > 20, 20, pw.this.raw_price)
    )
)
```

```python
pn.extension()
# Define custom Bokeh plot function for daily pricing
def price_plotter(source, title="Dynamic Parking Price"):
    fig = bokeh.plotting.figure(
        height=400,
        width=800,
        title=title,
        x_axis_type="datetime",
        x_axis_label="Date",
        y_axis_label="Price ($)"
    )
    fig.line("t", "price", source=source, line_width=2, color="navy")
    fig.circle("t", "price", source=source, size=6, color="red")
    return fig

# Bind enhanced model output to live plot using Pathway’s streaming .plot()
#viz = enhanced_daily_window.plot(price_plotter, sorting_col="t")

# Display using Panel
#pn.Column(viz).servable()
```

```python
# VISUALIZATION FOR MODEL 1
tabs_model1 = []
for lot_id in range(1, 15):
    filtered = enhanced_daily_window.filter(pw.this.lot == lot_id)
    viz = filtered.plot(lambda src: price_plotter(src, f"Model 1 - Lot {lot_id}"), sorting_col="t")
    tabs_model1.append((f"Lot {lot_id}", pn.Column(viz)))

pn.panel(pn.Tabs(*tabs_model1)).servable()
```

**Model 2 - Demand Based Price Function**

```python
import datetime

# Step 1: Daily tumbling window with full feature aggregation
model2_window = (
    data_with_time.windowby(
        pw.this.t,
        instance=(pw.this.day, pw.this.parking_lot_number),
        window=pw.temporal.tumbling(datetime.timedelta(days=1)),
        behavior=pw.temporal.exactly_once_behavior()
    )
    .reduce(
        t=pw.this._pw_window_end,

        occ_sum=pw.reducers.sum(pw.this.Occupancy),
        cap=pw.reducers.max(pw.this.Capacity),  # we include capacity now

        queue_sum=pw.reducers.sum(pw.this.QueueLength),
        traffic_sum=pw.reducers.sum(pw.this.traffic_condition_weighted),
        vehicle_weight_sum=pw.reducers.sum(pw.this.VehicleType_weighted),
        special_day_flag=pw.reducers.max(pw.this.IsSpecialDay),
        lot = pw.reducers.max(pw.this.parking_lot_number),
        count=pw.reducers.count(),
    )
)
```

```python
#Step 2: Compute averages
model2_window = model2_window.with_columns(
    occ_avg = pw.this.occ_sum / pw.this.count,
    queue_avg = pw.this.queue_sum / pw.this.count,
    traffic_avg = pw.this.traffic_sum / pw.this.count,
    vehicle_weight_avg = pw.this.vehicle_weight_sum / pw.this.count
)
```

```python
#Step 3: Compute Demand score
model2_window = model2_window.with_columns(
    demand = pw.if_else(
        pw.this.cap == 0, 0,
        (
            0.35 * (pw.this.occ_avg / pw.this.cap) +
            0.2 * (pw.this.queue_avg / 5.0) +
            0.15 * (pw.this.traffic_avg / 2.0) +
            0.2 * pw.this.vehicle_weight_avg +
            0.1 * pw.this.special_day_flag
        )
    )
)
```

```python
#Step 4: Compute raw price
model2_window = model2_window.with_columns(
    raw_price = 10 + 10 * pw.this.demand
)
```

```python
#Step 5: Clamp final price to $5-$20
model2_window = model2_window.with_columns(
    price = pw.if_else(
        pw.this.raw_price < 5, 5,
        pw.if_else(
            pw.this.raw_price > 20, 20,
            pw.this.raw_price
        )
    )
)
```

```python
# VISUALIZATION FOR MODEL 2
tabs_model2 = []
for lot_id in range(1, 15):
    filtered = model2_window.filter(pw.this.lot == lot_id)
    viz = filtered.plot(lambda src: price_plotter(src, f"Model 2 - Lot {lot_id}"), sorting_col="t")
    tabs_model2.append((f"Lot {lot_id}", pn.Column(viz)))

pn.panel(pn.Tabs(*tabs_model2)).servable()
```
